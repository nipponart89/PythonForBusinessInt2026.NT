Task 1,
We are asked to filter out invalid transactions where the price is zero or the quantity is negative.
For each remaining transaction, compute a new field total = price * quantity.
Group the valid transactions by region and compute the total revenue per region. Store the result in a dictionary.
Print the region with the highest total revenue and its value using an f-string.

```python
#The list of transactions
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
#Sub task A
#First a new list of dicitionaries with filtered transactions is created
filtered_transactions = []
#The For loop sets the limits for valid transactions
for transaction in transactions:
    if transaction["price"] > 0 and transaction["quantity"] >= 0:
        #The .append command adds transaction that meet limits set earlier to the new list.
        filtered_transactions.append(transaction)

#Sub task B 
#Add total to the transactions in valid_transactions
for transaction in filtered_transactions:
    transaction["total"] = transaction["price"] * transaction["quantity"]

#Grouping by region
#first a dictionary for revenue assosicated with regions is created
region_revenue = {}
#The For loops sets variables region and total as the region and total defined in the filtered list.
for transaction in filtered_transactions:
    region = transaction["region"]
    total = transaction["total"]
    #The if loop checks if the region allready has a value in the region_revenue direcotry if not sets value as 0 
    if region not in region_revenue:
        region_revenue[region] = 0
    #Adds the total to the current total for this region    
    region_revenue[region] += total
#create variable for the best region 
# .get returns the value of item with the specified key
best_region = max(region_revenue, key=region_revenue.get)
#create variable for the value of the best region
best_regionvalue = region_revenue[best_region]
#print results of task 1 
print("Valid transactions")
print(filtered_transactions)
print("\nRevenue by region:")
print(f"\nRegion with highest revenue:{best_region} ({best_regionvalue})")
```

Task 2 asks to create a function given a list of daily sales figures. 

```python
#Define the given list 
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
#Define the function for the summarized sales list, summarise_slaes is given sales list as argument to avoid overridign the orginial list name. 
def summarise_sales(sales_list):
    total = sum(sales_list)
    average = round(total / len(sales_list), 2)
    best_day = sales_list.index(max(sales_list)) + 1
    worst_day = sales_list.index(min(sales_list)) + 1
    #The for loop creates the counting logic for days with sales exceeding average
    above_average_count = 0
    for sale in sales_list:
        if sale > average:
            above_average_count += 1
    #Returns the values of the summarised sales function as a dictionary
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }
#sets result as the summary of the function summarise sales. 
result = summarise_sales(sales)
print(result)
```

d) Find the student with the highest grade using a loop (do not use max()). Print the student's name and grade.

```python
#Creates a list of directories with the student name, grade and attendance.
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
#Creates list of passing students
passing = []
#Defines a passing student as a studen with grade greater or equal to 50. 
for student in students:
    if student["grade"] >= 50:
        #Adds the passing student name to the list of passing students
        passing.append(student["name"])
#Prints the names of the passing students from the passing list
print("Passing students:", passing)
#Sets arguments name and grade as name and grade from the student list of dictionaries
for student in students:
    name = student["name"]
    grade = student["grade"]
    #creates if loop for printing the correct grade; excellent, pass and, fail. 
    if grade >= 80:
        print(f"{name}: Excellent")
    elif 50 <= grade <= 79:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")
#sets the low attendance count as 0
low_attendance = 0

for student in students:
    #if loop sets limits for what counts as low attendance
    if student["attendance"] < 0.80:
        #Adds 1 to the low attendace count if the students attendance was under 0.80
        low_attendance += 1
#Prints the results for the low attendace loop.
print("Students with attendance below 0.80:", low_attendance)
#Sets the current top student as index number 1 in the lsit of dictionaries
top_student = students[0]

for student in students:
    #if the students grade is higher than current top student set student as new top student.
    if student["grade"] > top_student["grade"]:
        top_student = student

print(f"Top student loop: {top_student['name']} ({top_student['grade']})")
#finds top student using MAX, the pythoin lambda function is used to create a samll anonymous function in a single line.
top_studentmax = max(students, key=lambda student: student["grade"])
print(f"Top student max: {top_studentmax['name']} ({top_studentmax['grade']})")
```

Task 4, 
requiered to create a class called Protfolio. With constructor provied. 
Asked to add the following methods to the class:

a) A method add_asset(ticker, shares, price_per_share) that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method total_value() that returns the total market value of the portfolio.

c) A method most_valuable() that returns the ticker of the asset with the highest total value (shares × price).

d) A __str__ method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

```python
#Provided constructor starts
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share} #Provided constructor ends
    #Subtask A,
    #Defines the method for adding assets with argument self, ticker, shares and price_per_share
    def add_asset(self, ticker, shares, price_per_share):
        #sets argument self.holdings ticker as a combination of shares and price per share.
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }
    #Subtsk b,
    #Define method for setting total value from argument self.
    def total_value(self):
        total = 0
        #For loop updates information for the ticker total in holdings by adding amount of shares and price per share
        for ticker, info in self.holdings.items():
            total += info["shares"] * info["price_per_share"]
        return total
    #Subtask C,
    #Define method for lookign up the most valuable ticker from self
    def most_valuable(self):
        most_valuable_ticker = None
        highest_value = 0
        #For loop updates information on ticker values for spesific assets in holdings.
        for ticker, info in self.holdings.items():
            asset_value = info["shares"] * info["price_per_share"]
            if asset_value > highest_value: #Check if current asset value is higher than current highest value
                highest_value = asset_value #if yes set new highest asset 
                most_valuable_ticker = ticker #sets argument most_valuable_ticker as ticker
        #Returns the most valuable ticker
        return most_valuable_ticker
    #Subtask D,
    #Defines method for returning the readable summary
    def __str__(self):
        return (
            f"Portfolio owner: {self.owner}, "
            f"number of assets: {len(self.holdings)}, " #len retunrs the lenght of an object
            f"total value: {self.total_value()}"
        )

portfolio = Portfolio("Jaakko")

portfolio.add_asset("NVID", 10, 180)
portfolio.add_asset("UPM", 5, 420)
portfolio.add_asset("KONE", 8, 250)

print(portfolio)
print("Total value:", portfolio.total_value())
print("Most valuable asset:", portfolio.most_valuable())
print("Holdings:", portfolio.holdings)
```

Task 5,
Given a list of order dates as strings and product names compute and print the number of days between the earliest and latest order. Use datetime module and parse each date string into a datetime.date

```python
#Importing the module
from datetime import datetime
#Creating order list
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
#Converting date strings to date objects. 
converted_dates = []
for order in orders:
    date_string = order[0] #sets date_string as the order with index 0 
    converted_date = datetime.strptime(date_string, "%Y-%m-%d").date() #sets converted_date as date string with format year month day as the date
    converted_dates.append(converted_date) #adds converted dates to the converted dates list.
#Calculating differense between the number of days of first and lasts order
first_order = min(converted_dates) #sets the first order as the smallest i.e. earliest date
last_order = max(converted_dates) #sets the last orders as the largest i.e. latest date 
difference = last_order - first_order #sets difference as difference between last and first order as a timedelta. Contains days, seconds and microseconds
print("Days between first and last order:", difference.days) 
    
```

for a given set of quearterly revenues in thousand of €.
 Compute and print the total annual revenue for each year (i.e., row sums). Use the axis parameter.

b) Compute the mean quarterly revenue across all years (i.e., column means).

c) Using broadcasting, apply a growth adjustment to the 2024 row by multiplying it element-wise with the array [1.05, 1.03, 1.08, 1.02]. Print the adjusted values.

d) Use boolean indexing to find and print all quarterly revenue values that exceed 250 (thousands of euros).

```python
#import numpy as np to enable using np.array
import numpy as np
#Given in task description
# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])  #Given information ends
#Calculating the annual revenue for each year
annual_revenue = revenue.sum(axis=1) #Setting annual revenue as the sum of revenues for the rows i.e. horisontal axis
print("Annual revenue per annum:", annual_revenue)
#Calculating the mean revenues for each quarter
mean_quarterly_revenue = revenue.mean(axis=0) #setting the mean revenue for the quarters i.e. vertical axis
print("Mean quarterly revenue:", mean_quarterly_revenue)
#Calculating adjusted revenue for the year 2024 
growth_factors = np.array([1.05, 1.03, 1.08, 1.02]) #seting the groft factors as an array with the given vaules. 
adjusted_2024 = revenue[2] * growth_factors #setting adjsuted_2024 as the revenue from index number 2 so the year 2024, !numpy indexes rows!
print("Adjusted 2024 revenue:", adjusted_2024)
#Calculating the revenue values above 250
high_revenue = revenue[revenue > 250] #Settign high_revenue as revenues in the array above 250
print("Quarterly revenue values above 250:", high_revenue)
```

Task 7, 
Usign given data frame: 
a) Add a new column salary_per_year_exp computed as salary / years_exp. Round to the nearest integer.

b) Sort the DataFrame by performance_score in descending order and print the top 3 employees.

c) Using .loc, select all employees in the Finance department with a performance score above 4.0. Print the result.

d) Rename the column years_exp to experience_years using the .rename() method.

e) Using .iloc, select and print the salary and performance score columns for the first four employees.

```python
#import pandas and define as pd
import pandas as pd
#Given data
data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
} #given data ends
dataframe = pd.DataFrame(data) #Creates dataframe with dictionary data
dataframe["salary_experience"] = (dataframe["salary"] / dataframe["years_exp"]).round() #adds a new column in the data frame .round rounds to nearest integer
#Creating top 3 
top_3 = dataframe.sort_values(by="performance_score", ascending=False).head(3) #sorts values in data frame in decending order. .head returns first 5 in data frame unless otherwise specified
print("Top 3 employees by performance score:")
print(top_3)
#Using .loc find finance employees with performance above 4
finance_top = dataframe.loc[ #.loc used to choose information from dataframe
    (dataframe["department"] == "Finance") & (dataframe["performance_score"] > 4.0) #choosing those from dataframe with department finance and performance score > 4.0
]
print("\nFinance employees with performance score above 4.0:")
print(finance_top)
#Renaming
dataframe = dataframe.rename(columns={"years_exp": "experience_years"}) # .rename allows to choose what column or row to rename. 
print("\nDataFrame after renaming column:")
print(dataframe)
#Using .iloc to select and print slary and performance score columns for first four
selected_data = dataframe.iloc[:4, [2, 4]] #index based selection. :4 (rows part) take info upto but not includign index 4. [2,4] chooses columns indexed 2 and 4 
print("\nSalary and performance score for first four employees:")
print(selected_data)
```

Task 8, 
Generate data using given script. 
then:
a) Create a DataFrame with columns: date, visitors, conversion_rate, and avg_order_value. Set date as the index.

b) Compute two new columns: conversions (= visitors × conversion_rate, rounded to the nearest integer) and daily_revenue (= conversions × avg_order_value)

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
#Creating dataframe
websitetraf_df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

websitetraf_df = websitetraf_df.set_index("date") #sets date as index, date becomes the marker for rows instead of normal index

websitetraf_df["conversions"] = (websitetraf_df["visitors"] * websitetraf_df["conversion_rate"]).round() #calculates conversion by multiplying visitors with conversion rate
websitetraf_df["daily_revenue"] = websitetraf_df["conversions"] * websitetraf_df["avg_order_value"] #calculates daily revenue by multiplying conversions with average order value

print(websitetraf_df)
```
