Fixing data issues with listings.csv, all the data is contained in the first cell of the csv. leaving all other columns empty. 

```python
import csv

input_file = "listings.csv"
output_file = "listings_clean.csv"
bad_rows_file = "listings_bad_rows.csv"

expected_cols = 61

good_rows = 0
bad_rows = 0

with open(input_file, "r", encoding="utf-8", newline="") as infile, \
     open(output_file, "w", encoding="utf-8", newline="") as outfile, \
     open(bad_rows_file, "w", encoding="utf-8", newline="") as badfile:

    writer = csv.writer(outfile)
    bad_writer = csv.writer(badfile)

    # header
    header_line = infile.readline().rstrip("\n")
    header = next(csv.reader([header_line]))
    writer.writerow(header)
    bad_writer.writerow(["raw_row"])

    for line in infile:
        line = line.rstrip("\n")

        if not line:
            continue

        parsed_row = None

        # Try 1: normal CSV row
        try:
            row1 = next(csv.reader([line]))
            if len(row1) == expected_cols:
                parsed_row = row1
        except:
            pass

        # Try 2: whole row stored as one quoted string
        if parsed_row is None:
            try:
                row2_outer = next(csv.reader([line]))
                if len(row2_outer) == 1:
                    row2 = next(csv.reader([row2_outer[0]]))
                    if len(row2) == expected_cols:
                        parsed_row = row2
            except:
                pass

        # Write good row or save bad one separately
        if parsed_row is not None:
            writer.writerow(parsed_row)
            good_rows += 1
        else:
            bad_writer.writerow([line])
            bad_rows += 1

print(f"Clean file saved as {output_file}")
print(f"Bad rows saved as {bad_rows_file}")
print(f"Good rows kept: {good_rows}")
print(f"Bad rows removed: {bad_rows}")
```
