```python
import pandas as pd

df = pd.read_csv("listings_clean.csv", low_memory=False)
print(df.shape)
df.head()
```

```python
from dotenv import load_dotenv
import os

# Load variables from .env file
load_dotenv()

# Access API key
api_key = os.getenv("API_KEY")

# Print only first 5 characters
print(api_key[:5] + "...")
```

Task 2,a 

```python
import requests #library used for sending reguests to APIs
tickers = ["META", "AAPL", "AMZN", "NFLX", "GOOG"] 

url = "http://api.marketstack.com/v1/eod" #The API endpoint for marketstack EOD data. 

params = {
    "access_key": api_key,   #sets api_key from .env file as acces key, needed for the API request to go trough 
    "symbols": ",".join(tickers), #Formats the ticker list to a string with comma separated values
    "date_from": "2026-03-01", 
    "date_to": "2026-03-31",
    "limit": 1000 #limits records returned to 1000 rows
}

response = requests.get(url, params=params) #sends request to the API with the paramters set above
data = response.json() #converts the API response from JASON format to pyhton dictionary
```

Task 2, b 

```python
#makes a pandas data frame out of the JASON data
df_stocks = pd.DataFrame(data["data"])
df_stocks.head()
```

Task 2, c 

```python
# Keep only the columns needed
df_close = df_stocks[["symbol", "date", "close"]].copy() #.copy creates a copy of thsi slice of the data frame

# Make sure date is treated as a date and sort by stock and day
df_close["date"] = pd.to_datetime(df_close["date"])
df_close = df_close.sort_values(["symbol", "date"])

# Summary statistics for closing prices by stock
summary_stats = df_close.groupby("symbol")["close"].describe()
print("Summary statistics for closing prices:")
print(summary_stats)

# Monthly return percentage for each stock
returns = df_close.groupby("symbol").apply(
    lambda x: ((x.iloc[-1]["close"] - x.iloc[0]["close"]) / x.iloc[0]["close"]) * 100
).reset_index(name="return_pct")

print("\nMonthly return percentages:")
print(returns)

# Best and worst performer
best_stock = returns.loc[returns["return_pct"].idxmax()]
worst_stock = returns.loc[returns["return_pct"].idxmin()]

print(f"\nBest performer in March 2026: {best_stock['symbol']} ({best_stock['return_pct']:.2f}%)")
print(f"Worst performer in March 2026: {worst_stock['symbol']} ({worst_stock['return_pct']:.2f}%)")
```

Task 3, a

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))

sns.lineplot(data=df_stocks, x="date", y="close", hue="symbol")

plt.title("Closing Prices of FAANG Stocks - March 2026")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend(title="Stock")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Task 3, b

```python
plt.figure(figsize=(8,5))

sns.barplot(data=returns, x="symbol", y="return_pct")

plt.title("Monthly Return Percentage - March 2026")
plt.xlabel("Stock")
plt.ylabel("Return (%)")

plt.show()
```

Task 3, c

```python
# Create a pivot table for the heatmap
volume_pivot = df_stocks.pivot(index="symbol", columns="date", values="volume")

# Plot heatmap
plt.figure(figsize=(12, 5))
ax = sns.heatmap(volume_pivot, cmap="YlGnBu")
cbar = ax.collections[0].colorbar
cbar.set_label("Trading Volume")


plt.title("Daily Trading Volume for FAANG Stocks - March 2026")
plt.xlabel("Date")
plt.ylabel("Stock")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
