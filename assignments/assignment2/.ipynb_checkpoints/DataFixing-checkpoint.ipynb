Fixing data issues with listings.csv, all the data is contained in the first cell of the csv. leaving all other columns empty. 

```python
import csv

input_file = "listings.csv"
output_file = "listings_clean.csv"

with open(input_file, "r", encoding="utf-8", newline="") as infile, \
     open(output_file, "w", encoding="utf-8", newline="") as outfile:

    reader = infile.readlines()

    # First line = header (already correct)
    header = reader[0].strip()
    outfile.write(header + "\n")

    # Process remaining rows
    for line in reader[1:]:
        line = line.strip()

        # Remove outer quotes around entire row
        if line.startswith('"') and line.endswith('"'):
            line = line[1:-1]

        # Parse correctly and re-write as proper CSV
        parsed = next(csv.reader([line]))
        writer = csv.writer(outfile)
        writer.writerow(parsed)

print("Clean file saved as listings_clean.csv")
```
